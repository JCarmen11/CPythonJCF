```python
print("Esta es una calculadora")
print("Elije el tipo de operación del menú")
print("1.- Suma")
print("2.- Resta")
print("3.- Multiplicar")
print("4.- Dividir")
resp=input("¿Qué quieres hacer? ")
#
n1=float(input("Escribe el primer número: "))
n2=float(input("Escribe el segundo número: "))
if resp=="Suma":
  print("{} + {} = {}".format(n1,n2,n1+n2))
elif resp=="Restar":
  print("{} - {} = {}".format(n1,n2,n1-n2))
elif resp=="Multiplicar":
  print("{} x {} = {}".format(n1,n2,n1*n2))
elif resp=="Dividir":
  print("{} / {} = {}".format(n1,n2,n1/n2))
else:
  print("Error")
```

```python
#range
for n in range(15): # hasta uno antes del número del range
  print(n)

print("---")
for n in range(1,16): # de tal número al tal número
  print(n)

print("---")
for n in range(0,22,3): # ahora irá de tal número a otro tal número pero en salto de x número
  print(n)
```

```python
#while

n=1
print("Tabla del 5")
while n <= 10: # mientras se cumpla la condición después del while
  print(n*5) # haz esto otro
  n = n+1 # y esto otro
print("Fin") #hasta que no se cumpla pasará a esta línea
```
